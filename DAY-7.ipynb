**LOADING DATA FILE:**

```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv("general_data.csv")
df
```

**FINDING IF THERE ARE ANY NULL VALUES:**

```python
df.isnull().sum()
```

**FILLING NULL VALUES WITH MEAN OF THE COLUMN:**

```python
df['NumCompaniesWorked'] = df['NumCompaniesWorked'].fillna(df['NumCompaniesWorked'].mean())
df.isnull().sum()
```

**CONVERTING CATEGORICAL VARIABLES TO NUMERICAL:**

```python
from sklearn import preprocessing
le=preprocessing.LabelEncoder()
df["Attrition"]= le.fit_transform(df["Attrition"])
df["BusinessTravel"]=le.fit_transform(df["BusinessTravel"])
df["Department"]=le.fit_transform(df["Department"])
df["EducationField"]=le.fit_transform(df["EducationField"])
df["Gender"]=le.fit_transform(df["Gender"])
df["MaritalStatus"]=le.fit_transform(df["MaritalStatus"])
df["JobRole"]=le.fit_transform(df["JobRole"])
df
```

**NEW DATAFRAMES BASED ON YES AND NO:**

```python
df_y = df[df['Attrition']==1]
df_n = df[df['Attrition']==0]
print("shape of employees who left=",df_y.shape)
print("shape of employees who stayed=",df_n.shape)

```

**PERFORMING NON PARAMETRIC TESTS:**

**MANNWHITNEYY TEST:**

```python
column=df[['Age','DistanceFromHome','MonthlyIncome',
       'NumCompaniesWorked', 'PercentSalaryHike', 'TotalWorkingYears', 'TrainingTimesLastYear',
       'YearsAtCompany', 'YearsSinceLastPromotion', 'YearsWithCurrManager']]
from scipy.stats import mannwhitneyu
def mtest(col):
    print(f"\n{col} vs Attrition: \n")
    print(f"H0:There is no siginificant difference between {col} in between people who stayed and people who leaved")
    print(f"H1:There is siginificant difference between {col} in between people who stayed and people who leaved")
    stats,p=mannwhitneyu(df_y[col],df_n[col])
    if p<0.05:
        print("Reject null hypothesis and accept alternative hypothesis")
        print(f"{col} in people for attrition yes and attrition no has significant difference of {p} ")
    else:
        print("accept null hypothesis ")
        print(f"{col} has no significant difference of {p} in attrition yes and attrition no people\n\n")
            
cols = list(column)
for col in cols:
         mtest(col)
```

**CHISQUARE TEST:**

```python
col11=df[['Gender','MaritalStatus','BusinessTravel','Department','Education','EducationField','JobLevel',
         'JobRole','TrainingTimesLastYear']]
def chi(column2):
    print(f"\n{column2} vs Attrition:\n")
    print(f"H0:There is no dependency between {column2} and Attrition")
    print(f"H1:There is dependency between {column2} and Attrition")
    from scipy.stats import chi2_contingency
    chitable=pd.crosstab(df[column2],df['Attrition'])
    print(chitable)
    stats, p, dof, expected = chi2_contingency(chitable)
    print("\nstats and p :\n",stats,p)
    if p>0.05:
        print(f"Accept null hypothesis\n There is no dependency between {column2} and Attrition ")
    else:
        print(f"Accept Alternative hypothesis\n There is dependency of {round(p,3)} between {column2} and Attrition\ng")
cols1 = list(col11)
for cols in cols1:
    chi(cols)        
```

**PARAMETRIC TEST**

**TWO SAMPLE INDEPENDENT T TEST**

```python
col22=df[['DistanceFromHome','MonthlyIncome','YearsAtCompany','YearsWithCurrManager']]
def ttest(column3):
        print(f"\n{column3} vs attrition:\n")
        print(f"H0:There is no significance difference in means of  {column3} in employees who stayed and who left")
        print(f"H1:There is  significance difference in means of  {column3} in employees who stayed and who left\n")
        print("stats and p values:")
        from scipy.stats import ttest_ind
        stats,p=ttest_ind(df_y[column3],df_n[column3])
        print(stats,p)
        if p<0.05:
             print(f"Reject null hypothesis\n There is significance difference of {round(p)} in {column3} of employees who stayed and who left")
        else:
             print(f"Accept null hypothesis\n there is no significance difference of {column3} in employee who stayed and left\n\n")
cols11 = list(col22)
for colso in cols11:
        ttest(colso)         
```
