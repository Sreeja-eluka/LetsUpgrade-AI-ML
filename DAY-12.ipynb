**PERFORMING MANNWHITNEY TEST ON ATTRITION VS OTHER COLUMNS:**

```python
column=df1[['Age','DistanceFromHome','MonthlyIncome',
       'NumCompaniesWorked', 'PercentSalaryHike', 'TotalWorkingYears', 'TrainingTimesLastYear',
       'YearsAtCompany', 'YearsSinceLastPromotion', 'YearsWithCurrManager']]
from scipy.stats import mannwhitneyu
def mtest(col):
    print(f"\n{col} vs Attrition: \n")
    print(f"H0:There is no siginificant difference between {col} in between people who stayed and people who leaved")
    print(f"H1:There is siginificant difference between {col} in between people who stayed and people who leaved")
    stats,p=mannwhitneyu(df_y[col],df_n[col])
    if p<0.05:
        print("Reject null hypothesis and accept alternative hypothesis")
        print(f"{col} in people for attrition yes and attrition no has significant difference of {p} ")
    else:
        print("accept null hypothesis ")
        print(f"{col} has no significant difference of {p} in attrition yes and attrition no people\n\n")
            
    

```

```python
cols = list(column)
for col in cols:
    mtest(col)
    
```

**CHISQUARE TEST**

```python
col1=df1[['Gender','MaritalStatus','BusinessTravel','Department','Education','EducationField','JobLevel',
         'JobRole','TrainingTimesLastYear']]
def chi(column2):
    print(f"{column2} vs Attrition:\n")
    print(f"H0:There is no dependency between {column2} and Attrition")
    print(f"H1:There is dependency between {column2} and Attrition")
    from scipy.stats import chi2_contingency
    chitable=pd.crosstab(df1[column2],df1['Attrition'])
    print(chitable)
    stats, p, dof, expected = chi2_contingency(chitable)
    print("\nstats and p :\n",stats,p)
    if p>0.05:
        print(f"Accept null hypothesis\n There is no dependency between {column2} and Attrition ")
    else:
        print(f"Accept Alternative hypothesis\n There is dependency of {p}between {column2} and Attrition")
cols1 = list(col1)
for cols in cols1:
    chi(cols)        
        
```

**KRUSKALWALLIS TEST**

```python
dffr = df_y.loc[df_y['BusinessTravel']==2]
dfrr = df_y.loc[df_y['BusinessTravel']==1]
dfnr = df_y.loc[df_y['BusinessTravel']==0]
from scipy.stats import kruskal
stats,p=kruskal(dffr['YearsAtCompany'],dfnr['YearsAtCompany'],dfrr['YearsAtCompany'])
print(stats,p)
if p>0.05:
    print("accept null hypothesis\n There is no signifant differnce in years at company between\n he employees who left and has travel history as frequently,never,rarely")
else:
    print(f"reject null hypothesis\n there is significant difference of {p} in years at company between the people who eft andd has travelhistory as frequently,never,rarely")
          
          
    


```

**TWO SAMPLE INDEPENDENT T TEST**

```python
print("H0:There is no significance difference in means of  Age in employees who stayed and who left\n")
print("H1:There is  significance difference in means of  Age in employees who stayed and who left\n")
from scipy.stats import ttest_ind
stats,p=ttest_ind(df_y['Age'],df_n['Age'])
print(stats,p)
if p<0.05:
    print(f"Reject null hypothesis\n There is significance difference of {round(p)} in age of employees who stayed and who left")
else:
    print(f"Accept null hypothesis\n there is no significance difference of age in employee who stayed and left")
    
```
