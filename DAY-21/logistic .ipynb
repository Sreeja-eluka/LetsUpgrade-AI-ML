```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
import numpy as np
data = pd.read_excel('Bank_Personal_Loan_Modelling.xlsx',sheet_name=1)
data.head()
```

```python
data=data.drop(columns=['ZIP Code','ID'])
data.drop_duplicates(inplace=True) 
data.head(2)

```

```python
Y=data[['Personal Loan']]
X =data[['Age', 'Experience', 'Income', 'Family', 'CCAvg', 'Education', 'Mortgage', 'Securities Account', 'CD Account', 'Online', 'CreditCard']]
```

```python
import statsmodels.api as sm
import math as m 
X1 = sm.add_constant(X)
logistic=sm.Logit(Y,X1)
```

```python
result=logistic.fit()
result.summary()
```


Inference<br>

When a bank wants to offer a loan, the following factors are highly significant:<br><br>
1.Annual income of the customer<br>
2.Family size of the customer<br>
3.Education Level. 1: Undergrad; 2: Graduate; 3: Advanced/Professional<br>
4.Does the customer have a certificate of deposit (CD) account with the bank?<br>
5.Does the customer use internet banking facilities?<br>
6.Does the customer use a credit card issued by UniversalBank<br><br>
The following factors are also significant to some extent:<br>
1.Avg. spending on credit cards per month<br>
2.Does the customer have a securities account with the bank?<br><br>
    
These factors do not have any significance on personal loasn:<br>
1.Age of the customer<br>
2.Experienceof customer<br>

```python
df = pd.read_csv('general_data.csv')
df.head(2)
```

```python
df.isnull().sum()
```

```python
df=df.dropna()
df.isnull().sum()
```

```python
df = df.drop(columns = ['EmployeeID', 'EmployeeCount', 'Over18', 'StandardHours', ])
```

```python
from sklearn import preprocessing
le=preprocessing.LabelEncoder()
df['Attrition']=le.fit_transform(df['Attrition'])
df['BusinessTravel'] = le.fit_transform(df['BusinessTravel'])
df['EducationField'] = le.fit_transform(df['EducationField'])
df['Gender'] = le.fit_transform(df['Gender'])
df['Department'] = le.fit_transform(df['Department'])
df['MaritalStatus'] = le.fit_transform(df['MaritalStatus'])
df['JobRole'] = le.fit_transform(df['JobRole'])
```

```python
df.head()
```

```python
y=df.Attrition
x = df.drop(columns = ['Attrition'])
```

```python
x1 = sm.add_constant(x)
logistic=sm.Logit(y,x1)
```

```python
result=logistic.fit()
result.summary()
```
