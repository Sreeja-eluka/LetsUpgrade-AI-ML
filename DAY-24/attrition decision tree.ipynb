```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn import tree
from sklearn import preprocessing

```

```python
df=pd.read_csv("general_data.csv")
df.head(2)
```

```python
df['NumCompaniesWorked'] = df['NumCompaniesWorked'].fillna(df['NumCompaniesWorked'].mean())
df.isnull().sum()
```

```python
df=df.dropna()
```

```python
df.isnull().sum()
```

```python
le=preprocessing.LabelEncoder()
df["Attrition"]= le.fit_transform(df["Attrition"])
df["BusinessTravel"]=le.fit_transform(df["BusinessTravel"])
df["Department"]=le.fit_transform(df["Department"])
df["EducationField"]=le.fit_transform(df["EducationField"])
df["Gender"]=le.fit_transform(df["Gender"])
df["MaritalStatus"]=le.fit_transform(df["MaritalStatus"])
df["JobRole"]=le.fit_transform(df["JobRole"])
df
```

**dropping irrelevant columns**

```python
df = df.drop(['EmployeeCount','EmployeeID', 'Over18','StandardHours'], axis=1)
df.head()
```

**splitting Dataset**

```python
y=df.iloc[:,1:2]
y.head(4)
```

```python
x=df.drop(columns=["Attrition"])
x.head(2)
```

**TRAINING THE MODEL USING DECISION TREE**

```python
tree_model=tree.DecisionTreeClassifier()
tree_model.fit(x,y)
```

```python
tree_model.score(x,y)
```

with respect to score the model seems overfitting as the accuracy is 100%

**HERE ATTRITION IS DEPENDENT VARIABLE TO FIND IMPORTANT INDEPENEDENT VARIABLE WE WILL PERFORM RANDOM FOREST**

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_model=RandomForestClassifier(n_estimators=1000,max_features=2,oob_score=True)
```

```python
rf_model.fit(x,y)
```

```python
rf_model.oob_score_
```

**FINDING IMPORTANT VARIABLES:**

```python
a = ['Age', 'BusinessTravel', 'Department', 'DistanceFromHome','Education', 'EducationField', 'Gender', 'JobLevel', 'JobRole',
       'MaritalStatus', 'MonthlyIncome', 'NumCompaniesWorked','PercentSalaryHike', 'StockOptionLevel', 'TotalWorkingYears',
       'TrainingTimesLastYear', 'YearsAtCompany', 'YearsSinceLastPromotion','YearsWithCurrManager']
for feature,imp in zip(a,rf_model.feature_importances_):
    print(feature,"\t\t :",imp)
```

**CONCLUSION:**<br>
from above we can see that <br>
Age,DistanceFromHome,MonthlyIncome,TotalworkingYears,YearsAtCompany,PercentSalaryHike has higher impact on Attrition

**TAKING INDEPENDENT VARIABLES WHICH ARE  IMPORTANT**

```python
imp=df[['Age','DistanceFromHome','MonthlyIncome','TotalWorkingYears','YearsAtCompany','PercentSalaryHike']]
imp.head(3)
```

**DESICION TREE:**

```python
model = tree.DecisionTreeClassifier(max_depth=12)
model.fit(imp,y)
```

**SCORE:**

```python
model.score(imp,y)

```

**score is optimum now!!**

```python
impf=["Age","DistanceFromHome","MonthlyIncome","TotalWorkingYears","YearsAtCompany","PercentSalaryHike"]
from sklearn.tree import export_text
r=export_text(model,feature_names=impf)
print(r)
```
