```python
import pandas as pd

df = pd.read_csv('general_data.csv')

df.head()
```

```python
df.dropna()
```

```python
df["Attrition"].replace(to_replace=("No","Yes"),value=(0,1),inplace=True)
df
```

**1.CORRELATION BETWEEN ATTRITION AND AGE**

```python
from scipy.stats import pearsonr
stats,p=pearsonr(df.Attrition,df.Age)
print(stats,p)
if(stats==0):
            print("no correlation")
elif(stats>0):
             print("positively correlated with each other")
else:
    print("negatively correlated to each other")
if(p<0.05):
    print(" Null hypothesis rejected")
else:
    print("Null hypothesis accepted")
import matplotlib.pyplot as plt
plt.scatter(df.Attrition,df.Age)


    
    
    
    
```
